```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("sales_data.csv")
data
```

```python
data.info()
```

```python

data['Price'] = data['Price'].fillna(data['Price'].mean())

# 2. Fill missing numerical values with the median
data['Units_Sold'] = data['Units_Sold'].fillna(data['Units_Sold'].median())

# 3. Fill missing numerical values with 0 (or any constant value)
data['Discount'] = data['Discount'].fillna(0)

# 4. Fill missing categorical values with the most frequent value (mode)
data['Category'] = data['Category'].fillna(data['Category'].mode()[0])


# 6. Fill missing values using forward fill or backward fill
data['Price'] = data['Price'].fillna(method='ffill') # Forward fill
data['Discount'] = data['Discount'].fillna(method='ffill')
data['Category'] = data['Category'].fillna(method='ffill')
data['Units_Sold'] = data['Units_Sold'].fillna(method='ffill')

data.to_csv('sales_data.csv', index=False)


```

```python
correlation = data_cleaned["Price"].corr(data_cleaned["Sales"])
print(f"Correlation between Price and Sales: {correlation}")
if correlation > 0:
    print("Higher prices are associated with more sales.")
elif correlation < 0:
    print("Higher prices are associated with fewer sales.")
else:
    print("There is no correlation between Price and sales.")

```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_cleaned["Price"], data_cleaned["Sales"], alpha=0.5, color='b')

# Add titles and labels
plt.title("Scatterplot of Price vs Sales", fontsize=14)
plt.xlabel("Price", fontsize=12)
plt.ylabel("Sales", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# Show the plot
plt.show()
```

```python

# Ensure the 'Date' column is in datetime format
data_cleaned['Date'] = pd.to_datetime(data_cleaned['Date'])

# Sort the data by Date to ensure proper time order
data_cleaned = data_cleaned.sort_values(by='Date')

# Group the data by 'Date' and calculate the sum of sales per day
sales_by_date = data_cleaned.groupby('Date')['Sales'].sum()

# Create a line plot of Sales over time
plt.figure(figsize=(10, 6))
plt.plot(sales_by_date.index, sales_by_date.values, color='b', marker='o')

# Add titles and labels
plt.title("Sales Trend Over Time", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Sales", fontsize=12)

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python

# Create a figure with two subplots for Price and Sales
plt.figure(figsize=(12, 6))

# Boxplot for Price
plt.subplot(1, 2, 1)
sns.boxplot(data=data_cleaned, x='Price', color='lightblue')
plt.title('Price Distribution')

# Boxplot for Sales
plt.subplot(1, 2, 2)
sns.boxplot(data=data_cleaned, x='Sales', color='lightgreen')
plt.title('Sales Distribution')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Define price bins (adjust these ranges as necessary)
bins = [0, 100, 200, 300, 400, 500, 1000]  # Example price ranges
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501+']  # Labels for the bins

# Create a new column for price range
data_cleaned['Price_Range'] = pd.cut(data_cleaned['Price'], bins=bins, labels=labels)

# Group by Price_Range and calculate the average sales for each range
sales_by_price_range = data_cleaned.groupby('Price_Range')['Sales'].mean()

# Display the results
print(sales_by_price_range)
```

```python
# Create the bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=sales_by_price_range.index, y=sales_by_price_range.values, color='lightblue')

# Add titles and labels
plt.title("Average Sales by Price Range", fontsize=14)
plt.xlabel("Price Range", fontsize=12)
plt.ylabel("Average Sales", fontsize=12)

# Display values on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():,.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=12, color='black', 
                xytext=(0, 5), textcoords='offset points')

# Show the plot
plt.show()
```

```python
# Ensure 'Date' is in datetime format
data_cleaned['Date'] = pd.to_datetime(data_cleaned['Date'])

# Extract Month and Day of the Week from Date
data_cleaned['Month'] = data_cleaned['Date'].dt.month
data_cleaned['Day_of_Week'] = data_cleaned['Date'].dt.day_name()

# Group by Month and calculate the average sales for each month
sales_by_month = data_cleaned.groupby('Month')['Sales'].mean()

# Group by Day of the Week and calculate the average sales for each day
sales_by_day = data_cleaned.groupby('Day_of_Week')['Sales'].mean()

# Plotting the results
plt.figure(figsize=(12, 6))

# Plot sales by month
plt.subplot(1, 2, 1)
ax1 = sales_by_month.plot(kind='bar', color='lightblue')

# Set x-tick labels to month names
ax1.set_xticklabels(pd.to_datetime(sales_by_month.index, format='%m').month_name(), rotation=45)

plt.title('Average Sales by Month')
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.grid(True, linestyle='--', alpha=0.5)


# Plot sales by day of the week
plt.subplot(1, 2, 2)
sales_by_day = sales_by_day[['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']]  # Sort days in order
ax2 = sales_by_day.plot(kind='bar', color='lightgreen')
plt.title('Average Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Sales')
plt.grid(True, linestyle='--', alpha=0.5)

# Show the plot
plt.tight_layout()
plt.show()
```

# what insights you can draw about relationship between product sales and sales volume

Analyzing the relationship between product prices and sales volume can provide valuable insights into how pricing impacts sales performance. Based on the correlation, trends, and distribution of the data, here are some key insights you can explore:

1. Price and Sales Correlation
Negative Correlation: If you find a negative correlation (e.g., around -0.5 or lower), it indicates that as the price increases, the sales volume tends to decrease. This could suggest that consumers are price-sensitive and prefer lower-priced products. It might also indicate that higher prices are associated with more premium or less frequently purchased items.
Positive Correlation: A positive correlation (e.g., close to 0.5 or higher) suggests that as the price increases, sales also increase. This might happen if higher-priced products are seen as more valuable or if the market is less sensitive to price increases, potentially due to higher income groups or luxury items.
No Correlation: A weak or zero correlation suggests there is no significant linear relationship between price and sales volume, meaning that price may not be a deciding factor in purchasing behavior for this product.
